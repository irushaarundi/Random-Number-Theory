Linear Congruential Generator

```python
def linear_congruential_generator(seed, a, c, m, n):
    numbers = [seed] # seed parameter use to get the same sequence of randon numbers in every time
    for _ in range(n-1):
        next_number = (a* numbers[-1] + c) % m
        numbers.append(next_number/m)
    return numbers
```

```python
# LCG parameters
seed = 42
a=16807
c = 0
m = 2147483647
n = 1000
```

```python
# Generate random numbers using LCG
random_numbers = linear_congruential_generator(seed, a, c, m, n)
```

```python
# Scatter plot
plt.scatter(range(1, n + 1), random_numbers, s=5, color='blue', alpha=0.7)
plt.title('Linear Congruential Generator Scatter Plot')
plt.xlabel('Iteration')
plt.ylabel('Random Number')
plt.show()
```

```python
import matplotlib.pyplot as plt

def linear_congruential_generator(a, c, m, x0, n):
    random_numbers = []
    x = x0
    for _ in range(n):
        x = (a * x + c) % m
        random_numbers.append(x / m)
    return random_numbers

# Parameters
a = 16807
c = 0
m = 2**31 - 1
x0 = 100
N = 1000

# Generate random numbers
random_numbers = linear_congruential_generator(a, c, m, x0, N)

# Scatterplot
plt.scatter(range(1, N+1), random_numbers, marker='.')
plt.title('Scatterplot of Random Numbers')
plt.xlabel('Index')
plt.ylabel('Random Number')
plt.show()

# Histogram
plt.hist(random_numbers, bins=20, edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Random Number')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

def Linear_Congruential_Generater(X0, a, c, m, N):
    random_num = [] 
    Xt=X0
    for _ in range(N):
        Xt = (a*Xt + c) % m
        random_num.append(Xt/m)
    return(random_num)
```
