```python
import matplotlib.pyplot as plt
```

```python
def MRG32k3a(X,Y,m,N):
    MRG32k3a_rand_num=[]
    for _ in range(N):
        Xt=(1403580*X[-2] - 810728*X[-3])%m[0]
        X.append(Xt)
        Yt=(527612*Y[-2] - 1370589*Y[-3])%m[1]
        Y.append(Yt)
        if Xt <= Yt:
            Ut=(Xt - Yt + m[0])/(m[0] + 1)
            MRG32k3a_rand_num.append(Ut)
        else:
            Ut=(Xt - Yt)/(m[0]+1)
            MRG32k3a_rand_num.append(Ut)
    return MRG32k3a_rand_num
```

```python
X=[1,2,3]
Y=[4,5,6]
m=[4294967087,4294944443]
N=1000
rand_num=MRG32k3a(X,Y,m,N)
print(rand_num)
#Scatter plot
plt.scatter(range(1,N+1),rand_num, s=5)
plt.xlabel("Iteration")
plt.ylabel("Random Numbers")
plt.show()
```

```python
#Histogram
plt.hist(rand_num, edgecolor="black",width=0.1)
plt.title("Histogram of MRG32k3a")
plt.xlabel("Random Number")
plt.ylabel("Frquency")
plt.show()
```
