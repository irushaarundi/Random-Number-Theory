```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
P=[]
for i in range(3):
    i+=1
    P.append(i/6)
```

```python
def TLM(U, aI):
    I=[]
    X=[]
    n=len(aI)
    for u in U:
        celin_val=np.ceil(n*u)
        I.append(celin_val)
    return I
```

```python
X0= 100
N=10**5
a = 16807
c = 0
m = 2147483647
U = P
aI = [1, 2, 2, 3, 3, 3]
I_values = TLM(U, aI)
mapped_values = [aI[int(i)-1] for i in I_values]
```

```python
plt.hist(mapped_values,edgecolor='black', bins=np.arange(1, 4) - 0.5) 
plt.xlabel('Mapped Values')
plt.ylabel('Frequency')
plt.title('Histogram of Mapped Values')
plt.xticks(range(1, 3))
plt.show()
```
