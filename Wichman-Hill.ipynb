```python
import matplotlib.pyplot as plt
```

```python
def Wichman_Hill(K, m,  N ):
    rand_num_wichH=[]
    Xt=K
    Yt=K
    Zt=K
    for _ in range(N):
        Xt=(171*Xt)%m[0]
        Yt=(172*Yt)%m[1]
        Zt=(170*Zt)%m[2]
        X=(Xt/m[0] + Yt/m[1] + Zt/m[2])%1
        rand_num_wichH.append(X)
    return rand_num_wichH
```

```python
K=100
m=[30269, 30307, 30323]
N=1000
Wichman_Hill_num = Wichman_Hill(K, m,  N )

#Scatter plot of Wichman_Hill
plt.scatter(range(1,N+1), Wichman_Hill_num,s=5)
plt.xlabel("Iteration")
plt.ylabel("Wichman_rand_number")
plt.title("Scatter plot of Wichman_Hill")
plt.show()
```

```python
#Histogram of Wichman_Hill
plt.hist(Wichman_Hill_num, edgecolor="black")
plt.xlabel("Withman_Hill_rand_num")
plt.ylabel("Frequence")
plt.title("Histogram of Wichman_Hill")
plt.show()
```
